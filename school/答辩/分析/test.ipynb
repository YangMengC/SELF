```python
import pandas as pd

file = "data.xls"
df = pd.read_excel(file) #reading file

#print(df)
data = df['日期'].values
#print(data)
level = df['质量等级'].values
aqi = df['AQI指数'].values
pm2 = df['PM2.5'].values
pm10 = df['PM10'].values
so2 = df['So2'].values
no2 = df['No2'].values
co = df['Co'].values
o3 = df['O3'].values
#print(o3)
```

```python
import matplotlib.pyplot as plt
def runplt():
    plt.figure()
    plt.grid(True)
    return plt
```

```python
plt1 = runplt()
plt1.title("the change of AQI")
plt1.plot(data, aqi, 'k.')
plt1.show()
```
